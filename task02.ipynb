```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression

# Sample dataset (no need to download)
data = {
    "text": [
        "payment failed issue",
        "unable to login account",
        "refund not received",
        "error in website loading",
        "billing issue with invoice",
        "password reset problem"
    ],
    "category": [
        "billing",
        "technical",
        "billing",
        "technical",
        "billing",
        "technical"
    ]
}

df = pd.DataFrame(data)

# Convert text to numbers
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(df['text'])

# Train model
model = LogisticRegression()
model.fit(X, df['category'])

# Test prediction
test = ["payment not working"]
test_vec = vectorizer.transform(test)
prediction = model.predict(test_vec)

print("Predicted Category:", prediction[0])
def assign_priority(text):
    if "failed" in text or "error" in text:
        return "High"
    elif "issue" in text:
        return "Medium"
    else:
        return "Low"

priority = assign_priority("payment not working")
print("Priority:", priority)
```
